```python
sarasai = []

with open('Salary_Data.csv') as failas:
    failas.readline()
    for eilute in failas:
        eilute = eilute.strip()
        isskaidyta = eilute.split(',')
        sarasas = dict(
           amzius = int(isskaidyta[0]),
           lytis = isskaidyta[1],
           issilavinimas = isskaidyta[2],
           pozicija = isskaidyta[3],
           patirtis = float(isskaidyta[4]),
           alga = int(isskaidyta[5])
         )
        
        #print(sarasas)
        sarasai.append(sarasas)

#print(sarasai[:5])

for sarasas in sarasai[:5]:
    print(sarasas)
```

```python
algos = [ sarasas['alga'] for sarasas in sarasai ]
print(algos)

vidurkis = round(sum(algos) / len(algos))
print('visų algų vidurkis:', vidurkis)

maziausia = min(algos)
print('mažiausia alga yra:', maziausia)

didziausia = max(algos)
print('dižiausia alga yra:', didziausia)
```

```python
patirtys = [ sarasas['patirtis'] for sarasas in sarasai ]
print(patirtys)

vidurkis = round(sum(patirtys) / len(patirtys))
print('visų darbuotojų darbo vidurkis (metais):', vidurkis)

maziausia = min(patirtys)
print('mažiausiai dirbta metų:', maziausia)

didziausia = max(patirtys)
print('Didžiausia patirtis (metais):', didziausia)
```
